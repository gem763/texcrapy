```python
from gensim.models import Doc2Vec
```

```python
d2v = Doc2Vec.load('model/doc2vec.model')
```

```python
d2v.wv.most_similar(positive=['nike'], negative=[], topn=100)
```

```python
d2v.wv.similarity('nike', 'iphone')
```

```python
vec = d2v.infer_vector(['아이폰','iphone','애플','apple'], epochs=500)
d2v.docvecs.most_similar(positive=[vec], topn=20)
```

```python
vec = d2v.infer_vector(['갤럭시','galaxy', '삼성','samsung'], epochs=500)
d2v.docvecs.most_similar(positive=[vec], topn=20)
```
