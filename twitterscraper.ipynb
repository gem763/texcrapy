```python
from twitterscraper import query_tweets
import pandas as pd
import json
```

```python
for i in xl.index[295:]:
    _xl = xl.iloc[i]
    qry = '#' + _xl['한글수정'] + ' OR #' + _xl['영문수정']
    title = _xl['영문수정']
    scrap(title, qry)
```

```python
def scrap(brand, qry, lang='ko'):
    #bd = pd.Timestamp('2014-01-01')
    ed = pd.Timestamp('2018-12-31').date()

    tw = []
    for t in query_tweets(qry, lang=lang, enddate=ed):
        tw.append({
            'id': t.id, 
            'date': t.timestamp, 
            'user': t.user, 
            'text': t.text, 
        })
    
    df = pd.DataFrame(tw).set_index('id')
    df.date = pd.to_datetime(df.date)
    df.to_pickle('socialdb/' + brand + '~2018.kr.pkl')
    return df
```
